# Diretrizes para Prompting
Nesta lição, você praticará dois princípios de prompting e suas táticas relacionadas para escrever prompts eficazes para grandes modelos de linguagem.

## Chave de API
Trabalharemos com o modelo do Google, o Gemini 2.5 Flash. As bibliotecas utilizadas encapsulam chamadas para a API do Google que utilizam esse modelo.

Uma chave para uso dessa API deve ser gerada. Ela identifica o usuário e controla o uso que é feita da API.

Acesse [Google AI Studio](https://aistudio.google.com/app/api-keys) para gerar essa API.

## Configuração
#### Carregue a chave da API e as bibliotecas Python relevantes.

````python
from google import genai
from google.colab import userdata

client = genai.Client(api_key=userdata.get('GOOGLE_API_KEY'))
````

````python
from google.genai import types

def get_completion(prompt):
    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=prompt,
        config=types.GenerateContentConfig(
            temperature=0,
            top_p=0.95,
            top_k=20,
        ),
    )
    return response.text
````

## Princípios de Prompting
- **Princípio 1: Escreva instruções claras e específicas**
- **Princípio 2: Dê tempo para o modelo "pensar"**

### Táticas

#### Tática 1: Use delimitadores para indicar claramente partes distintas da entrada
- Delimitadores podem ser qualquer coisa como: ```, """, < >, `<tag> </tag>`, `:`

````python
text = f"""
Você deve expressar o que deseja que um modelo faça \
fornecendo instruções que sejam tão claras e \
específicas quanto possível. \
Isso guiará o modelo em direção à saída desejada, \
e reduzirá as chances de receber respostas irrelevantes \
ou incorretas. Não confunda escrever um \
prompt claro com escrever um prompt curto. \
Em muitos casos, prompts mais longos fornecem mais clareza \
e contexto para o modelo, o que pode levar a \
saídas mais detalhadas e relevantes.
"""
prompt = f"""
Resuma o texto delimitado por três crases \
em uma única frase de até 20 palavras.
```{text}```
"""
response = get_completion(prompt)
print(response)
````

#### Tática 2: Peça uma saída estruturada
- JSON, HTML

````python
prompt = f"""
Gere uma lista de três títulos de livros inventados junto \
com seus autores e gêneros.
Forneça-os em formato JSON com as seguintes chaves:
book_id, title, author, genre.
"""
response = get_completion(prompt)
print(response)
````

#### Tática 3: Peça para o modelo verificar se as condições foram satisfeitas

````python
text_1 = f"""
Fazer uma xícara de chá é fácil! Primeiro, você precisa ferver \
água. Enquanto isso acontece, \
pegue uma xícara e coloque um saquinho de chá dentro. Assim que a água estiver \
quente o suficiente, basta despejá-la sobre o saquinho de chá. \
Deixe descansar um pouco para o chá infundir. Após \
alguns minutos, retire o saquinho de chá. Se você \
quiser, pode adicionar um pouco de açúcar ou leite a gosto. \
E é isso! Você tem uma deliciosa \
xícara de chá para saborear.
"""
prompt = f"""
Você receberá um texto delimitado por três aspas.
Se contiver uma sequência de instruções, \
reescreva essas instruções no seguinte formato:

Passo 1 - ...
Passo 2 - …
…
Passo N - …

Se o texto não contiver uma sequência de instruções, \
então simplesmente escreva "Nenhum passo fornecido."

\"\"\"{text_1}\"\"\"
"""
response = get_completion(prompt)
print("Conclusão para o Texto 1:")
print(response)
````

````python
text_2 = f"""
O sol está brilhando intensamente hoje, e os pássaros estão \
cantando. É um dia lindo para dar um \
passeio no parque. As flores estão florescendo, e as \
árvores estão balançando suavemente na brisa. As pessoas \
estão por aí, aproveitando o clima agradável. \
Alguns estão fazendo piqueniques, enquanto outros estão jogando \
ou simplesmente relaxando na grama. É um \
dia perfeito para passar tempo ao ar livre e apreciar a \
beleza da natureza.
"""
prompt = f"""
Você receberá um texto delimitado por três aspas.
Se contiver uma sequência de instruções, \
reescreva essas instruções no seguinte formato:

Passo 1 - ...
Passo 2 - …
…
Passo N - …

Se o texto não contiver uma sequência de instruções, \
então simplesmente escreva "Nenhum passo fornecido."

\"\"\"{text_2}\"\"\"
"""
response = get_completion(prompt)
print("Conclusão para o Texto 2:")
print(response)
````

#### Tática 4: Prompting com "few-shot"

````python
prompt = f"""
Sua tarefa é responder em um estilo consistente.

<child>: Ensine-me sobre paciência.

<grandparent>: O rio que esculpe o mais profundo \
vale flui de uma fonte modesta; a \
mais grandiosa sinfonia origina-se de uma única nota; \
a tapeçaria mais intrincada começa com um fio solitário.

<child>: Ensine-me sobre resiliência.
"""
response = get_completion(prompt)
print(response)
````

### Princípio 2: Dê tempo para o modelo "pensar"

#### Tática 1: Especifique os passos necessários para completar uma tarefa

````python
text = f"""
Em uma charmosa vila, os irmãos Jack e Jill partiram em \
uma missão para buscar água em um poço no topo da colina. \
Enquanto subiam, cantando alegremente, a desgraça \
atingiu—Jack tropeçou em uma pedra e rolou \
colina abaixo, com Jill seguindo-o. \
Embora um pouco machucados, a dupla voltou para casa com \
abraços reconfortantes. Apesar do contratempo, \
seus espíritos aventureiros permaneceram inabaláveis, e eles \
continuaram explorando com alegria.
"""
# exemplo 1
prompt_1 = f"""
Realize as seguintes ações:
1 - Resuma o seguinte texto delimitado por três \
crases em 1 frase.
2 - Traduza o resumo para o francês.
3 - Liste cada nome no resumo em francês.
4 - Saída de um objeto json que contenha as seguintes \
chaves: french_summary, num_names.

Separe suas respostas com quebras de linha.

Texto:
```{text}```
"""
response = get_completion(prompt_1)
print("Conclusão para o prompt 1:")
print(response)
````

#### Peça a saída em um formato especificado

````python
prompt_2 = f"""
Sua tarefa é realizar as seguintes ações:
1 - Resuma o seguinte texto delimitado por
  <> em 1 frase.
2 - Traduza o resumo para o francês.
3 - Liste cada nome no resumo em francês.
4 - Saída de um objeto json que contenha as
  seguintes chaves: french_summary, num_names.

Use o seguinte formato:
Texto: <texto a ser resumido>
Resumo: <resumo>
Tradução: <tradução do resumo>
Nomes: <lista de nomes no resumo>
Saída JSON: <json com resumo e num_names>

Texto: <{text}>
"""
response = get_completion(prompt_2)
print("\nConclusão para o prompt 2:")
print(response)
````

#### Tática 2: Instrua o modelo a elaborar sua própria solução antes de chegar a uma conclusão

````python
prompt = f"""
Determine se a solução do aluno está correta ou não.

Pergunta:
Estou construindo uma instalação de energia solar e preciso \
ajuda para calcular os custos.
- O terreno custa $100 / pé quadrado
- Posso comprar painéis solares por $250 / pé quadrado
- Negociei um contrato de manutenção que custará \
uma taxa fixa de $100 mil por ano, e um adicional de $10 / pé \
quadrado
Qual é o custo total para o primeiro ano de operações \
em função do número de pés quadrados.

Solução do Aluno:
Seja x o tamanho da instalação em pés quadrados.
Custos:
1. Custo do terreno: 100x
2. Custo do painel solar: 250x
3. Custo de manutenção: 100.000 + 100x
Custo total: 100x + 250x + 100.000 + 100x = 450x + 100.000
"""
response = get_completion(prompt)
print(response)
````

#### Observe que a solução do estudante na verdade não está correta.
#### Podemos corrigir isso instruindo o modelo a elaborar sua própria solução primeiro.

````python
prompt = f"""
Sua tarefa é determinar se a solução do aluno \
está correta ou não.
Para resolver o problema, faça o seguinte:
- Primeiro, elabore sua própria solução para o problema, incluindo o total final.
- Em seguida, compare sua solução com a solução do aluno \
e avalie se a solução do aluno está correta ou não.
Não decida se a solução do aluno está correta até
que você tenha resolvido o problema você mesmo.

Use o seguinte formato:
Pergunta:
```
pergunta aqui
```
Solução do aluno:
```
solução do aluno aqui
```
Solução real:
```
etapas para elaborar a solução e sua solução aqui
```
A solução do aluno é igual à solução real \
que você acabou de calcular:
```
sim ou não
```
Nota do aluno:
```
correta ou incorreta
```

Pergunta:
```
Estou construindo uma instalação de energia solar e preciso ajuda \
para calcular os custos.
- O terreno custa $100 / pé quadrado
- Posso comprar painéis solares por $250 / pé quadrado
- Negociei um contrato de manutenção que custará \
uma taxa fixa de $100 mil por ano, e um adicional de $10 / pé \
quadrado
Qual é o custo total para o primeiro ano de operações \
em função do número de pés quadrados.
```
Solução do aluno:
```
Seja x o tamanho da instalação em pés quadrados.
Custos:
1. Custo do terreno: 100x
2. Custo do painel solar: 250x
3. Custo de manutenção: 100.000 + 100x
Custo total: 100x + 250x + 100.000 + 100x = 450x + 100.000
```
Solução real:
"""
response = get_completion(prompt)
print(response)
````
