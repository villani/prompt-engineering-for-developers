## Configuração
#### Carregue a chave da API e as bibliotecas Python relevantes.

```python
from google import genai
from google.colab import userdata

client = genai.Client(api_key=userdata.get('GOOGLE_API_KEY'))
```

### Parâmetros do LLM

- `temperature`: Controla a aleatoriedade das respostas. Valores baixos (como 0) tornam as respostas mais determinísticas e focadas, enquanto valores altos aumentam a criatividade e variedade.
- `top_p`: Define a probabilidade acumulada para seleção das próximas palavras. Um valor de 0.95 significa que o modelo considera as palavras mais prováveis até atingir 95% de probabilidade, tornando as respostas mais diversificadas sem perder qualidade.
- `top_k`: Limita o número de opções de palavras que o modelo pode escolher a cada passo. Por exemplo, com top_k=20, apenas as 20 palavras mais prováveis são consideradas, ajudando a controlar a criatividade e a relevância das respostas.

```python
from google.genai import types

def get_completion(prompt, temperature=0, top_p=0.95, top_k=20):
    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=prompt,
        config=types.GenerateContentConfig(
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
        ),
    )
    return response.text
```

```python
prompt = "Resuma em 50 palavras sobre o que é crébito."
response = get_completion(prompt, 0, 0.95, 20)
print(response)
```

```python
prompt = "Resuma em 50 palavras sobre o que é crébito."
response = get_completion(prompt, 1, 0.1, 100)
print(response)
```

```python
prompt = "Resuma em 50 palavras sobre o que é crébito."
response = get_completion(prompt, 2, 0.95, 20)
print(response)
```

```python
prompt = "Resuma em 25 palavras sobre o que é juros."
response = get_completion(prompt, 0, 1, 20)
print(response)
```

```python
prompt = "Resuma em 25 palavras sobre o que é juros."
response = get_completion(prompt, 0, 0, 20)
print(response)
```

Deixar o modelo desregulado, com temperatura alta e top_p em 1.0, ou seja, sem filtro de probabilidade faz com que o resultado seja bem diferente do que se espera, com informações erradas e sem sentido. O modelo parece estar "alucinando" e inventando coisas, o que é um comportamento comum quando os parâmetros de geração são configurados para serem muito criativos.

```python
prompt = "Faça uma lista resumida das cinco características mais marcantes do CDI do Fatec Praia Grande."
response = get_completion(prompt, 2, 1, 100)
print(response)
```

```python
prompt = "Faça uma lista resumida das cinco características mais marcantes do CDI do Fatec Praia Grande."
response = get_completion(prompt, 0, 0, 10)
print(response)
```
